#### Langkah 1: Muat Dataset Breast Cancer

```python
# Import library yang diperlukan
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Muat dataset Breast Cancer
cancer = load_breast_cancer()

# Konversi ke DataFrame pandas
df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target
df['target'] = df['target'].apply(lambda x: cancer.target_names[x])

# Tampilkan 5 baris pertama dari DataFrame
df.head()
```

#### Langkah 2: Lakukan Eksplorasi Data Dasar

```python
# Cek nilai yang hilang
print(df.isnull().sum())

# Statistik deskriptif
print(df.describe())

# Cek distribusi dari target
print(df['target'].value_counts())
```

#### Visualize the Data

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pilih beberapa fitur utama untuk visualisasi
subset_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area']

# Sampling sebagian data untuk visualisasi (menggunakan sampel 100 data)
df_sample = df.sample(100, random_state=42)

# Visualisasi distribusi fitur-fitur menggunakan pairplot
sns.pairplot(df_sample[subset_features + ['target']], hue='target', markers=["o", "s"])
plt.show()

# Boxplot untuk membandingkan distribusi fitur antara target
plt.figure(figsize=(8, 6))  # Kurangi ukuran gambar
sns.boxplot(x='target', y='mean radius', data=df)
plt.show()
```
